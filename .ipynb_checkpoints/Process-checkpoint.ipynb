# Input

```python
from skimage.io import imread, imshow, imsave
import numpy as np
GFP = 'data/gfp_rscl.tif'
MYC = 'data/mcy_rscl.tif'
gfp=imread(GFP)
myc=imread(MYC)
gfp=gfp[1:10,:,:]
myc=myc[1:10,:,:]
```

```python
gfp=gfp.astype('float64')
myc=myc.astype('float64')
```

# Take average intensity

```python
ave=np.add(gfp,myc)/2
ave.shape
```

```python
import skimage.io as io
io.imshow(ave[1,:,:],cmap='Greys')
```

# gaussian filter

```python
import skimage.filters as filters
ave_gau=np.asarray(list(map(lambda x: filters.gaussian(ave[x,:,:],sigma=2), range(9))))
io.imshow(ave_gau[1,:,:])
```

```python
from skimage.morphology import disk
from skimage.filters import threshold_otsu, rank
from skimage.util import img_as_ubyte

ave = img_as_ubyte(ave)
radius = 45
selem = disk(radius)
local_otsu = np.asarray(list(map(lambda x: rank.otsu(ave[x,:,:], selem),range(9))))
```
