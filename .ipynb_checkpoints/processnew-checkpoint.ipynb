```python
from collections.abc import Iterable
from skimage.transform import integral_image
import itertools
```

```python

def _mean_std(image, w):
    if not isinstance(w, Iterable):
        w = (w,) * image.ndim
    #_validate_window_size(w)

    pad_width = tuple((k // 2 + 1, k // 2) for k in w)
    padded = np.pad(image.astype('float'), pad_width,
                    mode='reflect')
    padded_sq = padded * padded

    integral = integral_image(padded)
    integral_sq = integral_image(padded_sq)

    kern = np.zeros(tuple(k + 1 for k in w))
    for indices in itertools.product(*([[0, -1]] * image.ndim)):
        kern[indices] = (-1) ** (image.ndim % 2 != np.sum(indices) % 2)

    total_window_size = np.prod(w)
    sum_full = correlate_sparse(integral, kern, mode='valid') # to do
    m = sum_full / total_window_size
    sum_sq_full = correlate_sparse(integral_sq, kern, mode='valid')
    g2 = sum_sq_full / total_window_size
    # Note: we use np.clip because g2 is not guaranteed to be greater than
    # m*m when floating point error is considered
    s = np.sqrt(np.clip(g2 - m * m, 0, None))
    return m, s
```

```python
from skimage.util import dtype as dtype
```

```python
def threshold_sauvola(image, window_size=15, k=0.2, r=None):
    if r is None:
        imin, imax = dtype.dtype_limits(image, clip_negative=False)
        r = 0.5 * (imax - imin)
    m, s = _mean_std(image, window_size)
    return m * (1 + k * ((s / r) - 1))
```

```python
# self-implement sauvola
r = 9
rr=int((r-1)/2)
k=0.5
for i in range(1):
    img=ave[i,:,:]
    m=np.zeros(shape=(ave.shape[1]-2*rr,ave.shape[2]-2*rr))
    s=np.zeros(shape=(ave.shape[1]-2*rr,ave.shape[2]-2*rr))
    T=np.zeros(shape=(ave.shape[1]-2*rr,ave.shape[2]-2*rr))
    for a in range(rr,ave.shape[1]-rr):
        for b in range(rr,ave.shape[2]-rr):
            submatrix=img[a-rr:a+rr+1,b-rr:b+rr+1]
            matrix_mean=np.mean(submatrix)
            matrix_std=np.std(submatrix)
            m[a-rr,b-rr]=matrix_mean
            s[a-rr,b-rr]=matrix_std
    T=m*(1+k*(s/128-1))
    sauvola=img[rr:ave.shape[1]-rr,rr:ave.shape[2]-rr]>=T
    
```
